```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# Import crash data
file1 = "LA-Collision-Analysis/Resources/Crashes_2017.csv"
df1 = pd.read_csv(file1)
file2 = "LA-Collision-Analysis/Resources/Crashes_2018.csv"
df2 = pd.read_csv(file2)
file3 = "LA-Collision-Analysis/Resources/Crashes_2019.csv"
df3 = pd.read_csv(file3)
```

```python
# Concatenate the 3 datasets
crashes_df = pd.concat([df1, df2, df3])
crashes_df
```

```python
# Quick glance at data
crashes_df.info()
```

```python
# Turn all '-' values to NaN
crashes_df = crashes_df.replace('-', np.NaN, regex=True)
crashes_df.isnull().sum()
```

```python
# Checking out one specific column for unique values... values coded as letters - need legend or conversion
crashes_df["TYPE_OF_COLLISION"].value_counts()
```

## Exploring Coords

```python
# Looking at Lat/Lng values - checking for nulls
print(f'Null Lat values: {crashes_df["LATITUDE"].isnull().sum()}')
print(f'Null Lng values: {crashes_df["LONGITUDE"].isnull().sum()}')
        # Null X values: 4893
        # Null Y values: 4893

# Lat and Long have roughly same coords as X and Y, yet many are null values

# Looking at X/Y coord values - checking for nulls
print(f'Null X values: {crashes_df["POINT_X"].isnull().sum()}')
print(f'Null Y values: {crashes_df["POINT_Y"].isnull().sum()}')
        # Null Lat values: 114921
        # Null Lng values: 114921

# Would like to find a way to fill in Lat/Lng values where POINT X/Y values are empty
```

## Dropping columns

```python
# dropping columns then exploring remainder
crashes_df_cleaned = crashes_df.drop(columns=[
    "PROC_DATE", "JURIS", "COLLISION_DATE", "OFFICER_ID",
    "REPORTING_DISTRICT", "CHP_SHIFT", "BEAT_TYPE", "CHP_BEAT_TYPE",
    "CITY_DIVISION_LAPD", "CHP_BEAT_CLASS", "BEAT_NUMBER", "PRIMARY_RD",
    "SECONDARY_RD", "DISTANCE", "DIRECTION", "INTERSECTION", "CALTRANS_COUNTY",
    "CALTRANS_DISTRICT", "STATE_ROUTE", "ROUTE_SUFFIX", "POSTMILE_PREFIX", "POSTMILE",
    "RAMP_INTERSECTION", "NOT_PRIVATE_PROPERTY", "PRIMARY_RAMP", "SECONDARY_RAMP", "COUNTY",
    "PCF_CODE_OF_VIOL", "PCF_VIOLATION", "PCF_VIOL_SUBSECTION", "SIDE_OF_HWY", "CNTY_CITY_LOC",
    "WEATHER_2", "CITY", "PED_ACTION", "CHP_VEHTYPE_AT_FAULT", "SPECIAL_COND",
    "STATE_HWY_IND", "CHP_ROAD_TYPE", "ROAD_COND_2", "LOCATION_TYPE", "POPULATION",
    "LONGITUDE", "LATITUDE"
    
])

crashes_df_cleaned
```

```python
# examining remaining columns
crashes_df_cleaned.info()
```

```python
# checking unique values per feature
crashes_df_cleaned.nunique()
```

```python
# Checking for null values across remaining dataset
crashes_df_cleaned.isnull().sum()
```

## Replacing binary values

```python
# Confirm columns with only 1 value count are "Y"
print(crashes_df_cleaned["PEDESTRIAN_ACCIDENT"].value_counts())
print(crashes_df_cleaned["BICYCLE_ACCIDENT"].value_counts())
print(crashes_df_cleaned["MOTORCYCLE_ACCIDENT"].value_counts())
print(crashes_df_cleaned["TRUCK_ACCIDENT"].value_counts())
print(crashes_df_cleaned["ALCOHOL_INVOLVED"].value_counts())
```

```python
# Replace null value with 0, and "Y" with 1
crashes_df_cleaned["PEDESTRIAN_ACCIDENT"].fillna(0, inplace=True)
crashes_df_cleaned["PEDESTRIAN_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df_cleaned["BICYCLE_ACCIDENT"].fillna(0, inplace=True)
crashes_df_cleaned["BICYCLE_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df_cleaned["MOTORCYCLE_ACCIDENT"].fillna(0, inplace=True)
crashes_df_cleaned["MOTORCYCLE_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df_cleaned["TRUCK_ACCIDENT"].fillna(0, inplace=True)
crashes_df_cleaned["TRUCK_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df_cleaned["ALCOHOL_INVOLVED"].fillna(0, inplace=True)
crashes_df_cleaned["ALCOHOL_INVOLVED"].replace("Y", 1,inplace=True)
```

```python
# Replace other null values with 0
crashes_df_cleaned["STWD_VEHTYPE_AT_FAULT"].value_counts()
crashes_df_cleaned["STWD_VEHTYPE_AT_FAULT"].fillna(0, inplace=True)
```

```python
# Checking null values again
crashes_df_cleaned.isnull().sum()
```

## Dropping null values

```python
crashes_df_cleaned.dropna(inplace=True)
crashes_df_cleaned
```

## Converting Values

```python
# convert floats to ints (aside from X/Y and ACCIDENT_YEAR, which needs datetime)
dict_column_datatypes = {"DAY_OF_WEEK" : int, "COLLISION_SEVERITY" : int, "NUMBER_KILLED" : int, 
                         "NUMBER_INJURED" : int, "PARTY_COUNT" : int, "COUNT_SEVERE_INJ" : int,
                         "COUNT_VISIBLE_INJ" : int, "COUNT_COMPLAINT_PAIN" : int, "COUNT_PED_KILLED" : int,
                         "COUNT_PED_INJURED" : int, "COUNT_BICYCLIST_KILLED" : int, "COUNT_BICYCLIST_INJURED" : int, 
                         "COUNT_MC_KILLED" : int, "COUNT_MC_INJURED" : int, "COLLISION_TIME" : int}
crashes_df_cleaned = crashes_df_cleaned.astype(dict_column_datatypes)
crashes_df_cleaned.dtypes
```

```python
crashes_df_cleaned["ACCIDENT_YEAR"].astype(str)
```

```python
# Convert ACCIDENT_YEAR to datetime
crashes_df_cleaned["ACCIDENT_YEAR"] = pd.to_datetime(crashes_df_cleaned["ACCIDENT_YEAR"], format='%Y')
crashes_df_cleaned["ACCIDENT_YEAR"]
```

```python
# Remove month and day - only show year
crashes_df_cleaned["ACCIDENT_YEAR"] = crashes_df_cleaned["ACCIDENT_YEAR"].dt.year

# Converts back to int64????
crashes_df_cleaned["ACCIDENT_YEAR"]
```

## Export File

```python
# Export file
crashes_df_cleaned.to_csv("Resources/crash_cleaned.csv", index=False)
```
